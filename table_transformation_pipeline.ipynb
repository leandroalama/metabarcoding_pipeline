```python
df_counts=clas_df.groupby(['taxid','barcode']).size().unstack(fill_value=0)
df_counts=df_counts.drop(['unclassified'], axis=1)
df_counts=df_counts.drop([-1], axis=0)
df_counts = df_counts.rename(columns=new_names)
df_counts.index.name = None
df_counts
```
